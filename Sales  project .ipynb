# Exploratory Data Analysis

### Gender

```python
 #width is used to change width and saturation is used to light /dark colour
var=sns.countplot(x="Gender", data=fs,hue="Gender",width=0.5,palette="Oranges_r") 
for bars in var.containers:
    var.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount

sales_gen = fs.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen,palette="Oranges_r",hue="Gender",width=.5)
plt.title("Plotting a bar chart for gender vs total amount",fontsize=20)
```

```python
# plotting a bar chart for gender vs total amount
sales_gen = (
    fs.groupby('Gender', as_index=False)['Amount']
      .sum()
      .sort_values(by='Amount', ascending=False)
)

plt.figure(figsize=(6,4))

var = sns.barplot(x='Gender', y='Amount', data=sales_gen,hue="Gender",width=.5,saturation=.9,palette="Oranges_r")

for bars in var.containers:
    var.bar_label(bars, fmt='%.0f')

plt.title("Total Sales by Gender")
plt.show()
```

##### Above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

### Age

```python
var=sns.countplot(data=fs,x="Age Group",hue="Gender",palette ="Oranges_r")
for bars in var.containers:
    var.bar_label(bars,fmt="%.0f")
plt.show()
```

```python
#Total Amount vs Age Group
sales_gen=(fs.groupby(["Age Group"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False))
plt.figure(figsize=(6,4))
var = sns.barplot(x='Age Group', y='Amount', data=sales_gen,hue="Age Group",width=.7,palette="Oranges_r")

for bars in var.containers:
    var.bar_label(bars, fmt='%.0f')

plt.title("Total Amount vs Age Group",fontsize=20)
plt.show()
```

##### From above graphs we can see that most of the buyers are of age group between 26-35 yrs female

### State

```python
plt.figure(figsize=(18,7))
var = sns.barplot(x='State', y='Orders', data=sales_gen,hue="State",width=.6,saturation=.9,palette="Accent")

for bars in var.containers:
    var.bar_label(bars, fmt='%.0f')

plt.title("Total number of orders from top 10 states",fontsize=20)
plt.show()
```

```python
#Total amount/sales from top 10 states
sales_gen=(fs.groupby(["State"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)).head(10)
plt.figure(figsize=(18,7))
var = sns.barplot(x='State', y='Amount', data=sales_gen,hue="State",width=.6,saturation=.7,palette="Accent")

for bars in var.containers:
    var.bar_label(bars, fmt='%.0f')

plt.title("Total amount/sales from top 10 states",fontsize=20)
plt.show()
```

##### From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

### Marital Status

```python
plt.figure(figsize=(6,4))
var = sns.countplot(data=fs, x='Marital_Status',hue="Marital_Status",width=.5,saturation=.7,palette="Oranges")

for bars in var.containers:
    var.bar_label(bars, fmt='%.0f')

plt.title("Marital Status",fontsize=20)
```

### Product Category

```python
plt.figure(figsize=(26,6))
var = sns.countplot(data=fs, x='Product_Category',hue='Product_Category',width=.8,saturation=.5,palette="gist_rainbow_r")

for bars in var.containers:
    var.bar_label(bars, fmt='%.0f')

plt.title('Product_Category',fontsize=20)
```

```python
#Most Sold Product
sales_gen=(fs.groupby(['Product_Category'], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)).head(10)
plt.figure(figsize=(20,7))
var = sns.barplot(x='Product_Category', y='Amount', data=sales_gen,hue='Product_Category',width=.6,saturation=.5,palette="Accent")

for bars in var.containers:
    var.bar_label(bars, fmt='%.0f')

plt.title("Most Sold Product",fontsize=20)
plt.show()
```

##### From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

```python
sales_gen=(fs.groupby(['Product_ID'], as_index=False)["Orders"].sum().sort_values(by="Orders", ascending=False)).head(10)
plt.figure(figsize=(20,7))
var = sns.barplot(x='Product_ID', y='Orders', data=sales_gen,hue='Product_ID',width=.6,palette="Oranges_r")

for bars in var.containers:
    var.bar_label(bars, fmt='%.0f')

plt.title("Most Sold Product",fontsize=20)
plt.show()
```

### Top 10 Sold Product

```python
sales_top10 = (
    fs.groupby(['Product_ID'], as_index=False)['Orders']
      .sum()                            # total orders per product
      .sort_values(by='Orders', ascending=False)  # sort descending
      .head(10)                         # take top 10
)

plt.figure(figsize=(16,6))

var = sns.barplot(
    x='Product_ID',
    y='Orders',
    data=sales_top10,
    hue='Product_ID',       # gives different color per product
    width=0.6,
    palette="Accent"     # reversed oranges palette
)

# Add values on top of bars
for bars in var.containers:
    var.bar_label(bars, fmt='%.0f')

plt.title("Top 10 Most Sold Products", fontsize=20)
plt.xlabel("Product ID", fontsize=12)
plt.ylabel("Total Orders", fontsize=12)      
plt.show()

```
